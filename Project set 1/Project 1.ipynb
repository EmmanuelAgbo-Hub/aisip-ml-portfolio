### A program that calculates exchange rates between 3 African currencies using variables and operators. 
###### AUTHOR: AGBO EMMANUEL SUNDAY

#### The three African countries of interest are 
* Ghana
* South Africa
* Nigeria

```python
# Declaring the conversion factors

# 1 Ghana cedi(GHS) = 122.16 Nigerian Naira(NGN)
GHS_TO_NGN = 122.16

# 1 South African Rand(ZAR) = 0.675 Ghana Cedi(GHS)
ZAR_TO_GHS = 0.675

# 1 Nigerian Naira(NGN) = 0.012 South African Rand(ZAR)
NGN_TO_ZAR = 0.012

# NOTE: All exchange rates are rates as at 13th of APRIL 2026 

naira = float(input('Enter amount to convert from Nigerian Naira to South African Rand: '))
print(f'{naira} NGN is equal to {round(naira*NGN_TO_ZAR, 2)} ZAR')

ghs = float(input('Enter amount to convert from Ghana Cedi to Nigerian Naira: '))
print(f'{ghs} GHS is equal to {round(ghs*GHS_TO_NGN, 2)} NGN')

zar = float(input('Enter amount to convert from South African Rand(ZAR) to Ghana Cedi(GHS): '))
print(f'{zar} ZAR is equal to {round(zar*ZAR_TO_GHS, 2)} GHS')
```

### INSIGHT
Among the 3 countries of interest, <br>
Ghana cedi ranks highest in value<br>
South African Rand follows <br>
Nigerian Naira ranks the lowest
