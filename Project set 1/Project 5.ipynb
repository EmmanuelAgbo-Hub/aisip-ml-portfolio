#### A program using a function that takes a list of student scores and returns the average, highest, and lowest using tuple unpacking.
#### AUTHOR: AGBO EMMANUEL SUNDAY

```python
def score_analysis(scores):
    average = round(sum(scores)/len(scores), 2)
    highest = max(scores)
    lowest = min(scores)
    return average, highest, lowest
```

```python
math_scores = [5, 4,6, 10, 12, 11, 13, 3, 5, 12, 10]
average, highest, lowest = score_analysis(math_scores)
```

```python
print(f"The average score is {average}")
print(f"The highest score is {highest}")
print(f"The lowest score is {lowest}")
```
